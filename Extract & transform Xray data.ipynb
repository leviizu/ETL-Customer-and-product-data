```python
# imorting packages
import pandas as pd
import json
```

```python
filename ='/Users/levy/Downloads/raw_xray_data.txt'
```

```python
# importing file
df=pd.read_table(filename, header=None, delimiter=None)[0].to_list()
```

```python
# converting json list to dictionary
data = json.loads(df[0])
```

```python
'''json_normalize fucntion isn't trnsforming json list properly, will perform manual transformation'''
def transform_data(data):
    # application
    application = data['application']

    # flip
    flip=data['captures'][0]['contexts'][0]['b-color']['flip']

    # roi
    roi_bottom = data['captures'][0]['contexts'][0]['b-color']['roi']['bottom']
    roi_horizontal_offset = data['captures'][0]['contexts'][0]['b-color']['roi']['horizontal offset']
    roi_horizontal_offset_top = data['captures'][0]['contexts'][0]['b-color']['roi']['top']

    # segmentation opacity
    capture_segmentation_opacity = data['captures'][0]['contexts'][0]['b-color']['segmentation opacity']

    # zoom
    zoom = data['captures'][0]['contexts'][0]['b-color']['zoom']

    # cine
    cine_date_and_time = data['captures'][0]['contexts'][0]['cine']['date and time']
    cine_id = data['captures'][0]['contexts'][0]['cine']['id']

    # cine frame
    cine_frame_end_timestamp = data['captures'][0]['contexts'][0]['cine frame']['end timestamp']
    cine_frame_start_timestamp = data['captures'][0]['contexts'][0]['cine frame']['start timestamp']

    # cine_range
    cine_range_from_hardware_id = data['captures'][0]['contexts'][0]['cine range']['from']['hardware id']
    cine_range_from_timestamp = data['captures'][0]['contexts'][0]['cine range']['from']['timestamp']
    cine_range_to_hardware_id = data['captures'][0]['contexts'][0]['cine range']['to']['hardware id']
    cine_range_to_timestamp = data['captures'][0]['contexts'][0]['cine range']['to']['timestamp']

    # images
    images_hardware_id = data['captures'][0]['contexts'][0]['images'][0]['hardware id']
    images_timestamp = data['captures'][0]['contexts'][0]['images'][0]['timestamp']

    # type
    types = images_timestamp = data['captures'][0]['contexts'][0]['type']
    # height
    height = data['captures'][0]['contexts'][0]['rect']['height']
    # width
    width = data['captures'][0]['contexts'][0]['rect']['width']
    # x
    rect_x = data['captures'][0]['contexts'][0]['rect']['x']
    # y
    rect_y = data['captures'][0]['contexts'][0]['rect']['y']
    # capture_date_and_time
    date_and_time = data['captures'][0]['date and time']

    # device
    device_contros_on_left = data['captures'][0]['device']['controls on left']
    device_pixels_per_mm = data['captures'][0]['device']['pixels per mm']
    device_screen_height = data['captures'][0]['device']['screen height']
    device_screen_width = data['captures'][0]['device']['screen width']

    # id
    capture_id = data['captures'][0]['id']

    #colection
    colection_tag = data['collections'][0]['tag']
    colection_title = data['collections'][0]['title']

    # date and time
    date_time = data['date and time']

    #impression
    impressions = data['impressions']

    # indications
    indications = data['indications']

    # institution
    institution = data['institution']

    # get study date
    study_date =  data['patient']['dates']['study date']

    # get patient age
    patient_age = data['patient']['quantities']['age']

    # institution_address=data
    institution_address=data['patient']['strings']['institution address']

    # institution_name=data
    institution_name=data['patient']['strings']['institution name']

    # operator_first_name
    operator_first_name=data['patient']['strings']['operator first name']

    # station_name
    station_name=data['patient']['strings']['station name']

    # get probe model
    probe_model = data['probe model']

    # get probe serial
    probe_serial = data['probe serial']

    # get study_id
    study_id = data['study id']

    # get workflow id
    workflow_id = data['workflow id']
    
    #adding all fields into a dictionary
    dict1 = {
        'application' : application,
        'flip': flip,
        'segmentation opacity': capture_segmentation_opacity,
        'roi_bottom': roi_bottom,
        'roi_horizontal_offset': roi_horizontal_offset,
        'roi_horizontal_offset_top': roi_horizontal_offset_top,
        'cine_date_and_time': cine_date_and_time,
        'cine_id': cine_id,
        'cine_frame_end_timestamp': cine_frame_end_timestamp,
        'cine_frame_start_timestamp': cine_frame_start_timestamp,       
        'cine_range_from_hardware_id' : cine_range_from_hardware_id,
        'cine_range_from_timestamp' : cine_range_from_timestamp,
        'cine_range_to_hardware_id' : cine_range_to_hardware_id,
        'cine_range_to_timestamp' : cine_range_to_timestamp, 
        'images_hardware_id' : images_hardware_id,
        'images_timestamp' : images_timestamp,        
        'types' : types,
        'height' : height, 
        'width' : width,
        'rect_x' : rect_x,
        'rect_y' : rect_y,    
        'date_and_time' : date_and_time,
        'device_contros_on_left': device_contros_on_left, 
        'device_pixels_per_mm' : device_pixels_per_mm,
        'device_screen_height': device_screen_height, 
        'device_screen_width' : device_screen_width, 
        'capture_id' : capture_id,       
        'colection_tag' : colection_tag,
        'colection_title' : colection_title,
        'date_time' : date_time,
        'impressions': impressions,
        'indications' : indications,
        'institution' : institution,
        'study_date =' : study_date,
        'patient_age' : patient_age,
        'institution_address' : institution_address,
        'institution_name' : institution_name,
        'operator_first_name' : operator_first_name,
        'station_name' : station_name,
        'probe_model' : probe_model,
        'probe_serial' : probe_serial,
        'study_id' : study_id,
        'workflow_id' : workflow_id}
    return dict1
```

```python
df = pd.DataFrame()
data=transform_data(data)
clarius_data = df.append(data, ignore_index=True)
```
